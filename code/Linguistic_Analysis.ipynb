- \# utterances

```python
import pandas as pd
import numpy as np
import nltk
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
from nltk import pos_tag
from nltk.corpus import stopwords
from scipy.stats import ttest_ind
from nltk.tokenize import word_tokenize
import string
import sys
import os


import pickle
path_subtl='/content/drive/My Drive/Colab Notebooks/Data/ADRC/data/'
stop_words = stopwords.words("english")


def get_word_dist(input_frame, trans_col):
    """
    get the log word frequency distrubtion on the input frame,
    :param pandas.DataFrame input_frame: the input frame with transcripts to evalute
    :param str trans_col: the column name represents transcript text
    :return: the log frequency for the transcript tokens
    :rtype: dict
    """
    word_dist = load_word_dist()
    log_lf = {}
    trans = input_frame[trans_col].values.tolist()
    for tran in trans:
        try:
            words = tran.split(" ")
            for word in set(words):
              # log_lf[word.lower()]={}
                # freq=get_word_lf(word, word_dist)
                # for key in freq:
                log_lf[word.lower()] = get_word_lf(word, word_dist)
        except AttributeError:
            continue
    # log_lf = sorted(log_lf.items(), key=lambda item: item[1])
    return log_lf


def get_word_lf(token, word_dist):
    """
    get the log lexical frequency for a specifc word
    :param str token: the word for calcualtion
    :param dict word_dict: the dictionary for the word raw frequency
    :return: the log lexical frequency for the word
    :rtype: float
    """
    if token.lower() in word_dist:
      return word_dist[token.lower()]
    else:
      return -1


def load_word_dist(option=1):
    """
    load Subtlex.US.text file for estimating word frequency distribution
    save it to local file
    return the word frequency distribution
    :rtype: nltk.probability.FreqDist
    """
    #option=0, log freq, 1 means count freq.
    if os.path.exists(path_subtl+"word_dist.pkl"):
        with open(path_subtl+"word_dist.pkl", "rb") as f:
            word_dist = pickle.load(f)
        return word_dist
    else:
        word_dist = {}
        sys.stdout.write("estimating frequency distribution...\n")
        df=pd.read_excel(path_subtl+'SUBTLEXusfrequencyabove1.xlsx')
        for idx,row in df.iterrows():




          word=str(row['Word']).lower()


          if word not in word_dist:
            #  word_dist[word]={}

            #  word_dist[word]['Lg10WF']=row['Lg10WF'] #log freq.
            #  word_dist[word]['SUBTLWF']=row['SUBTLWF'] #freq count

            if option==0:
              word_dist[word]=row['Lg10WF'] #log freq.
            else:
              word_dist[word]=row['SUBTLWF'] #freq count


        sys.stdout.write("done\n")
        # with open(path+"word_dist.pkl", "wb") as f:
        #     pickle.dump(word_dist, f)
        # print(len(word_dist))
        return word_dist
def calculate_lexical_frequency(tokens,tit):
    """
    calculate the lexical frequency for control & dementia model output,
    run t-test on two list
    :param con_tokens: a list of pre-processed control model output tokens
    :type con_tokens: list
    :param dem_tokens: a list of pre-processed dementia model output tokens
    :type dem_tokens: list
    """

    word_dist = load_word_dist()
    lf={}
    # calculate the occurrence of negative infinity
    print(tit)
    for key in tokens:
      print('%s model'%key)
      con_lf = [get_word_lf(token, word_dist) for token in tokens[key]]
      con_lf = [ele for ele in con_lf if ele != -1]
      con_ni_count = np.count_nonzero(np.isneginf(con_lf))
      lf[key] = [token for token in con_lf if token != -np.inf]
      con_lf=lf[key]
      print("Number of tokens removed from model: {}".format(con_ni_count))
      print("model mean log lexical frequency: {}".format(round(sum(con_lf)/len(con_lf), 2)))
      print("Control model unique word ratio: {:0.2f}".format(len(set(tokens[key]))/len(tokens[key])))
    keys=list(lf.keys())
    for i in range(0,len(keys)-1):
      for j in range(i+1,len(keys)):
        print('t test %s vs. %s'%(keys[i],keys[j]))


        # if p<0.05 -> significant difference between two samples
        print("t-test p-value: {:0.3f}".format(ttest_ind(lf[keys[i]], lf[keys[j]], equal_var=False)[1]))
def pre_process(res_df,col,tit):
    """
    add more stopwords a list of generated text,
    remove stopwords from the generate text
    return the cleaned tokens
    :param res_df: the dataframe for langauge generation
    :type res_df: pd.DataFrame
    :return: control model tokens and dementia model tokens
    :rtype: list, list
    """
    distribution={}
    temp={}
    tokens={}

    ####con vs. ad vs. mci
    if 'ncc' in col:
      distribution['con'] = res_df[res_df["ncc"]==0]['clean_utt'].values.tolist()
      distribution['dem'] = res_df[res_df["ncc"]==2]['clean_utt'].values.tolist()
      distribution['mci'] = res_df[res_df["ncc"]==1]['clean_utt'].values.tolist()
    elif 'label' in col:

      distribution['con_bin'] = res_df[res_df["label"]==0]['clean_utt'].values.tolist()
      distribution['dem_bin'] = res_df[res_df["label"]==1]['clean_utt'].values.tolist()
    elif 'MOCA_impairment' in col:
      distribution['moca_con'] = res_df[res_df["MOCA_impairment"]==0]['clean_utt'].values.tolist()
      distribution['moca_imp'] = res_df[res_df["MOCA_impairment"]==1]['clean_utt'].values.tolist()
    #accumulating stopwords
    for key in distribution.keys():
          # print(distribution[key][0])
          con_tokens = word_tokenize(distribution[key][0])
          con_tokens = [token.lower() for token in con_tokens if token not in string.punctuation]
          con_temp = [token for token in con_tokens if token.startswith("'")]
          stop_words.extend(con_temp)
          tagged = pos_tag(con_tokens)
          for item in tagged:
              tag = item[1]
              token = item[0]
              if tag in ("PRP", "PRP$", "WP$", "EX"):
                  stop_words.append(token)
          temp[key]=con_tokens

    for key in temp.keys():
        # print(temp[key])

         temps= [token for token in temp[key] if token not in stop_words]
         print(len(temps))
         tokens[key]=temps

    calculate_lexical_frequency(tokens,tit)

def parcipant_lf(utt):
  word_dist = load_word_dist()
  con_tokens = word_tokenize(utt)
  con_tokens = [token.lower() for token in con_tokens if token not in string.punctuation]
  con_temp = [token for token in con_tokens if token.startswith("'")]
  stop_words.extend(con_temp)
  tagged = pos_tag(con_tokens)
  for item in tagged:
      tag = item[1]
      token = item[0]
      if tag in ("PRP", "PRP$", "WP$", "EX"):
          stop_words.append(token)
  temps= [token for token in con_tokens if token not in stop_words]
  con_lf = [get_word_lf(token, word_dist) for token in temps]
  con_lf = [ele for ele in con_lf if ele != -1]
  con_lf= [token for token in con_lf if token != -np.inf]
  return round(sum(con_lf)/len(con_lf), 2)


measure='SUBTLscore'

for t in participant_utterances:
  print(t)

  if len(participant_utterances[t])==0:
    print('no utt. in file '+str(t))
    data[t][measure + " (participant)"]=0.0
    continue
  data[t][measure + " (participant)"]=parcipant_lf(participant_utterances[t])




# df=pd.read_csv(path+'utterance_data.csv')
# option=1 #0 is preprocess file, 1 is calculate lf
# if option==0:
#   column=['label','MOCA_impairment','ncc']
#   title=['con vs. dem', 'con vs. MOCA_impairment','con vs. mci vs. dem']
#   for idx,col in enumerate(column):
#     pre_process(df,col,title[idx])
# else:
#   parcipant_lf(df)
```

# Extract Linguistic Information
I'm currently extracting part-of-speech (POS) tags at a relatively course-grained level, following the Universal Dependencies scheme: https://universaldependencies.org/u/pos/.  If additional lexical and/or grammatical properties are needed, I can easily switch to a more complex scheme. ðŸ™‚

I'm extracting named entity labels using a model trained on the OntoNotes 5 corpus: https://catalog.ldc.upenn.edu/LDC2013T19.  This model produces the following named entity types (reproduced from the spaCy documentation: https://spacy.io/api/annotation#named-entities):


TYPE|DESCRIPTION
--- | ---
PERSON|People, including fictional.
NORP|Nationalities or religious or political groups.
FAC|Buildings, airports, highways, bridges, etc.
ORG|Companies, agencies, institutions, etc.
GPE|Countries, cities, states.
LOC|Non-GPE locations, mountain ranges, bodies of water.
PRODUCT|Objects, vehicles, foods, etc. (Not services.)
EVENT|Named hurricanes, battles, wars, sports events, etc.
WORK_OF_ART|Titles of books, songs, etc.
LAW|Named documents made into laws.
LANGUAGE|Any named language.
DATE|Absolute or relative dates or periods.
TIME|Times smaller than a day.
PERCENT|Percentage, including â€%â€œ.
MONEY|Monetary values, including unit.
QUANTITY|Measurements, as of weight or distance.
ORDINAL|â€œfirstâ€, â€œsecondâ€, etc.

# Compute Vocabulary Richness Features
I'm currently computing the following measures of syntactic complexity:


*   **Type-Token Ratio (TTR):** A well-known measure of lexical richness, computed as the number of _types_ (unique vocabulary words) divided by the number of _tokens_ (all instances of vocabulary words) in the text.  Standard TTR can be impacted by document length; the more complex measures of lexical richness below address this in different ways.
*   **Mean Segmental TTR:** Computes the average TTR across successive segments of text (e.g., words 1-100, 101-200, etc.), each of which contain the same number of tokens [(Johnson, 1944)](https://insights.ovid.com/pmga/194456020/00006828-194456020-00001).
*   **Moving-Average TTR:** Computes the average TTR across segments of text extracted using a moving window (e.g., words 1-100, 2-101, etc.) [(Covington and McFall, 2010)](https://www.tandfonline.com/doi/abs/10.1080/09296171003643098).
*   **Maas:** Computes a logarithmic correction of TTR [(Maas, 1972)](http://proxy.cc.uic.edu/login?url=https://search.proquest.com/docview/1300208330?accountid=14552).
*   **Measure of Textual Lexical Diversity:** Computes the TTR for increasingly longer segments of text.  When a predetermined threshold is reached, increments a counter and then restarts.  When the end of the document is reached, the number of tokens is divided by the final counter value.  The text sample is reversed and this process is repeated; the final measure is then computed by averaging the two scores (the one from the original text sample and the one from the reversed version) [(McCarthy, 2005)](http://proxy.cc.uic.edu/login?url=https://search.proquest.com/docview/305349212?accountid=14552).
*   **HD-D:** Computes a hypergeometric distribution of word occurrence across the document, and then sums the probabilities of each word type occurring in a random sample [(McCarthy and Jarvis, 2007)](https://journals.sagepub.com/doi/10.1177/0265532207080767).
